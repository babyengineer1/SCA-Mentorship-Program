# Question_4

```python
# all Rock music listener
```

```python
genre.rename(columns={'Name':'Genrename'}, inplace=True) # renaming the "Name" to "Genrename"
```

```python
genre.head(2)
```

```python
track_and_genre = pd.merge(track,genre)
```

```python
track_and_genre.head(2)
```

```python
invoiceline_and_invoice = pd.merge(invoiceline,invoice)
```

```python
invoiceline_and_invoice.head(2)
```

```python
new_merge_df = pd.merge(track_and_genre,invoiceline_and_invoice )
```

```python
new_merge_df.head(2)
```

```python
new_merge_df1 = pd.merge(new_merge_df,customer)
```

```python
new_merge_df1.head(2)
```

```python
new_merge_df2 = new_merge_df1.drop_duplicates(subset='Email', keep='first', inplace=False)
```

```python
genre_rock = new_merge_df2.loc[(new_merge_df1['Genrename']=='Rock'),['Email','FirstName','LastName','Genrename']]
```

```python
genre_rock.sort_values(['Email'], ascending = True).head(3)
```

# Question_5

```python
#most popular music genre
```

```python
spent = pd.DataFrame(new_merge_df1.groupby(['BillingCountry', "Genrename"], as_index=False)['UnitPrice'].sum())
spent.head()
```

```python
spent_in_total = pd.DataFrame(spent.groupby('BillingCountry', as_index=False)['UnitPrice'].max())
spent_in_total.head()
```

```python
pd.merge(spent_in_total, spent, on = ['UnitPrice', 'BillingCountry'], how = 'inner')
```
